```python
# ============================================================
# FINAL TESLIM ZIP OLUŞTURMA
# ============================================================

from pathlib import Path
import zipfile
import os

PROJECT_DIR = Path.cwd()

# Notebook adın buysa böyle kalsın.
# Farklıysa aşağıdaki dosya adını değiştir.
NOTEBOOK_FILE = PROJECT_DIR / "proje.ipynb"

LIGHT_ZIP_PATH = PROJECT_DIR / "UrbanFlow_AI_Final_Teslim.zip"

files_to_include = []

# Ana notebook
if NOTEBOOK_FILE.exists():
    files_to_include.append(NOTEBOOK_FILE)
else:
    print("[UYARI] Notebook dosyası bulunamadı:", NOTEBOOK_FILE)
    print("Mevcut .ipynb dosyaları:")
    for p in PROJECT_DIR.glob("*.ipynb"):
        print("-", p.name)

# Veri kaynakları dosyası
veri_kaynaklari = PROJECT_DIR / "veri_kaynaklari.txt"
if veri_kaynaklari.exists():
    files_to_include.append(veri_kaynaklari)

# Figures klasörü
FIGURES_DIR = PROJECT_DIR / "figures"
if FIGURES_DIR.exists():
    for p in FIGURES_DIR.rglob("*"):
        if p.is_file():
            files_to_include.append(p)

# Model outputs klasörü
OUTPUTS_DIR = PROJECT_DIR / "model_outputs"
if OUTPUTS_DIR.exists():
    for p in OUTPUTS_DIR.rglob("*"):
        if p.is_file():
            files_to_include.append(p)

# ZIP oluştur
with zipfile.ZipFile(LIGHT_ZIP_PATH, "w", zipfile.ZIP_DEFLATED) as zipf:
    for file_path in files_to_include:
        arcname = file_path.relative_to(PROJECT_DIR)
        zipf.write(file_path, arcname)

print("[OK] Teslim ZIP oluşturuldu:")
print(LIGHT_ZIP_PATH)

print("\nZIP içine eklenen dosyalar:")
for f in files_to_include:
    print("-", f.relative_to(PROJECT_DIR))

print("\nZIP boyutu:", round(LIGHT_ZIP_PATH.stat().st_size / (1024 * 1024), 2), "MB")
```

```python
from pathlib import Path
import zipfile

ROOT = Path.cwd()

ZIP_PATH = ROOT / "UrbanFlow_AI_Final_Teslim.zip"

# ZIP'e eklenecek ana dosya ve klasörler
items_to_zip = [
    ROOT / "proje.ipynb",
    ROOT / "veri_kaynaklari.txt",
    ROOT / "figures",
    ROOT / "model_outputs"
]

# Eski zip varsa sil
if ZIP_PATH.exists():
    ZIP_PATH.unlink()

with zipfile.ZipFile(ZIP_PATH, "w", zipfile.ZIP_DEFLATED) as zipf:
    for item in items_to_zip:
        if item.is_file():
            zipf.write(item, item.name)
            print("[FILE]", item.name)

        elif item.is_dir():
            for file_path in item.rglob("*"):
                if file_path.is_file():
                    arcname = file_path.relative_to(ROOT)
                    zipf.write(file_path, arcname)
                    print("[DIR FILE]", arcname)

        else:
            print("[UYARI] Bulunamadı:", item)

print("\n[OK] ZIP oluşturuldu:")
print(ZIP_PATH)

print("\nZIP boyutu:", round(ZIP_PATH.stat().st_size / (1024 * 1024), 2), "MB")

print("\nZIP içeriği:")
with zipfile.ZipFile(ZIP_PATH, "r") as z:
    for name in z.namelist():
        print("-", name)
```
